# 5 If Statement

if <condition>:
    <code>
elif <condition>:
    <code>
...
else:
    <code>

```python
age=18

if age>=18:
    print("You're an adult")

else:
    print("You're a kid")
```

# 6 For Loop

for <variable> in <iterable>:
    <code>

```python
countries
```

```python
for country in countries:
    print(country)
```

```python
#enumerate() function adds a counter to each item in a list or any other iterable
for i,country in enumerate(countries):
    print(i)
    print(country)
```

```python
my_data={"name":"kabir","age":20}
my_data
```

```python
for key,value in my_data.items():
    print(key)
    print(value)
```

```python
for i in range(1,10,2):
    print(i)
```
