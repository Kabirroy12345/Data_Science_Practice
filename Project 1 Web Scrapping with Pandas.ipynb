# Read .csv from Multiple URLs with Pandas

https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv

```python
import pandas as pd
import zipfile
import io
import requests
```

```python
df_iris = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
df_iris.head()
```

```python
urls = {
    "iris": "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv",
    "tips": "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv",
    "penguins": "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
}
```

```python
frames = []

for name in urls:
    df = pd.read_csv(urls[name])
    df.insert(0, "dataset", name)   # add column to identify dataset
    frames.append(df)

len(frames)
```

```python
frames[2].head(3)
```

```python
#Combine multiple CSVs
df_all = pd.concat(frames, ignore_index=True)
df_all.head()
```
