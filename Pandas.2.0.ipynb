# 1 Value Counts

```python
#len function
len(df_exams['gender'])
#.count method 
df_exams['gender'].count()
```

```python
#counting gender elements by category
df_exams['gender'].value_counts()
```

```python
#return the relative frequency(percentage) devide all values by the sum of values
df_exams['gender'].value_counts(normalize=True)
```

```python
#counting "parental level of education" elements by category
df_exams['parental level of education'].value_counts()
```

```python
#To get the relative frequency and round to 2 decimal places
df_exams['parental level of education'].value_counts(normalize=True).round(2)
```

# 1.Sort a dataframe

```python
#sort by one column
df_exams.sort_values('math score')
```

```python
# sort descending by one column(by default ascending=True)
df_exams.sort_values('math score',ascending=False)
```

```python
# sort descending by multiple columns
df_exams.sort_values(['math score','reading score','writing score'],ascending=False)
```

```python
# sort descending by multiple columns and update dataframe
#if i run df_exams there will be no change to make a change permanently
df_exams.sort_values(['math score','reading score','writing score'],ascending=False,inplace=True)
```

```python
# lets put race/ethinicity in descending order
df_exams.sort_values('race/ethnicity',ascending=True)
```

# 2. Create an Index

```python
import numpy as np
import random
```

```python
#creating non repetitive values for the index
new_index=np.arange(0,1000)
```

```python
#shuffling the indexes
random.shuffle(new_index)
```

```python
df_exams['new_index']= new_index
```

```python
df_exams.set_index('new_index',inplace=True)
df_exams
```

```python
df_exams.sort_index(ascending=True,inplace=True)
```

```python
df_exams
```
